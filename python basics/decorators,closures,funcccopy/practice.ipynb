```python
# iterators
# iterators are used as they dont load all the data at once using very less memory .
# they load and store one element at a time. and iterators show data in sequential order not all at once.
# next() is used for sequential printing 
# to create an iterator we use iter().
```

```python
lst=[1,2,3,4,5,6]
iterator=iter(lst) # list type iterator
print(type(iterator))
# after printing all the elements then printing again will show a stoperror which means elements finished to handle that use try except.
```

```python
try:
    print(next(iterator))
except StopIteration:
    print("stop iteration no elements left.") # after again and again running at last this exception will invoke.
else:
    print("no error found:") # this else run until no exception found.
finally:
    print("execcution complete") # this run no matter what condition
```

```python

# generators have iterators() and next() in them and is made through a function which have yeild instead of return
def my_gen():
    yield 1
    yield 2
    yield 3
for i in my_gen(): ## generator function object created not executed the function or sent any value there.
    print(i) # we do need to store the value returned if we print manually without loop so so with only loop we dont care about next()
#like this
gen=my_gen() ## so it creates my_gen() object gen but it doesn't execute anything until next(gen) called. thats how 1 print. first
for i in range(2):
    print(next(gen))
```

```python
def square(n):
    for i in range(n): ## yes it pauses where it was left and resume from same place. as for each number we need seperate yield as shown above with 3 yeild
        yield i**2
for i in square(11): ## generator function object created not executed the function or sent any value there.
    print(i)
# but if we handle manually than we have to store the yeild answer by our self
gen=square(3) ## generator function object created not excuted the generator function
for i in range(3):
    print(next(gen)) # if we directly write print(next(sqaure(3)) this will every time start from fresh we need to store answer of yeild in variable than give that to next
```

```python
# BUT MAINLY generators are used to read large file with ease. like
def read_large_files(file_path):
  with open(file_path,mode='r') as file:
    for line in file:
      yield line

file_path='large_file.txt'
for val in read_large_files(file_path): ## this doesn't send the file_path to generator function it only created the generator object. when print runs means the generaotr funciton is executed the first time.
  print(val.strip()) # .strip remove the \n after each line . that also prints if we remove strip
```

```python
# decorators # without calling any function the output is displayed.
def msg(func): 
    def sub_msg():
        func()
        print("this is a sub function that can also be created inside a function.")
    return sub_msg()
@msg # using @ sign and defining the funtion that will be printed inside of sub function here.
def say_hello(): #say_hello is the func name that is sent in sub_msg.
    print("hello")
```
