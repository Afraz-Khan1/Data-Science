Memory Management Best Practices:
1) Use local variables : local variables have short lifespan and are freed sooner than global variables.
2) avoid circular references: circular reference can lead to memory leaks. if not properly managed.
3) Use generators: generators produce the data from the dataset only when needed using yield. in this way less ram used.
4) delete objects explicitly: use the del statement to delete objects and variables.
5) use memory profiling tools like tracemalloc and memory_profiler to identify memory leaks and optimize memory usage.

```python
## handling circular reference with __del__ method. which arises problem for python to auto do garbage collection meomry free.
import gc
class myobj:
    def __init__(self,name): ## constructor. self is the object itself.
        self.name=name   ## initializing name
        print(f"objet {self.name} created")

    def __del__(self):
        print(f"Object {self.name} is deleted: ")
    
obj1=myobj("obj1")
obj2=myobj("obj2")
obj1.ref=obj2
obj2.ref=obj1

del obj1
del obj2

## manually trigger garbage collection. when objects have __del__ method. python can't safely destroy them auto so they are kept in memory for manual removal.
gc.collect() ## this collector imediately trigger the gc.garbage to free the unreachable memory like in cycles.
```

```python
##gc.collect() → actively runs the garbage collector to free unreachable objects immediately.
##gc.garbage → list of objects the collector couldn’t free (usually due to __del__ in cycles) for you to inspect or clean manually.

## which means this gc.collect auto runs garbage collector for all unreachable and other cyclic reference stuck memory to be freed
## but manually cyclic reference memory should be freed where __del__ exist or else python auto frees cyclic reference with no __del__ method and all other unreachable
# objects memory. 
```

```python
# so generators only bring data one at a time from dataset or forloop traversing in a range. so that less memory is used.
def generate_numbers(n):
    for i in range(1,n+1):
        yield i

size=int(input("Enter range: "))
for i in generate_numbers(size):
    print(i)
```
