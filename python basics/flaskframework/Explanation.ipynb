It is a web framework built using python programming language
Two importatnt things in flask framework.
1) WSGI (web server gateway interface.) it connects the web server from where user search web pages or request data, with the python application so that the app returns the answer for the user request. WSGI=bride between web server and python app
python application include both backend logic(routing,processing,data retrival on demand) and frontend.
2) jinja 2 template engine. used to upload or manupilate data to webpage from app backend. makes the pages dynamic.
