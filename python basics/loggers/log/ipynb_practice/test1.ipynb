```python
# loggers are used with end to end projects inorder to keep track of every error occurred at which position
import logging

logging.basicConfig(level=logging.DEBUG)
# log messages
logging.debug("this is a debug message")
logging.error("this is an error message")
logging.info("this is a info message")
logging.critical("this is a critical message")
logging.warning("this is a warning message")
```

```python
# configuring the logging 
import logging
logging.basicConfig(
    # in order to write these logs in file
    filename='app1.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s -%(name)s - %(levelname)s -%(message)s',
    datefmt='%Y-%m-%d %H-%M-%S'
)
 # now basic config is very necessary to do before using logger.
logging.debug("this is a debug message")
logging.info("this is a info message")
logging.error("this is an error message")
logging.warning("this is a warning message")
logging.critical("this is a critical message")
```
