```python
# logging with multiple loggers.
# we can use multiple loggers on different part of our code so that we can easily track which part have the error in the program.
```

```python
import logging
logger1=logging.getLogger("module1") # for example this is first part of the program and error here can ocure in some cases so we used our first logger here named module1.
logger1.setLevel(logging.DEBUG)

logger2=logging.getLogger("module2")
logger2.setLevel(logging.WARNING)

logging.basicConfig( # filename and filemode is added when needed.
    level=logging.DEBUG,# We use basicConfig(level=logging.DEBUG) to show all debug and higher-level log messages. like warning critical etc
    format='%(asctime)s - %(name)s - %(levelname)s -%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)
```

```python
logger1.debug("this is a debug message")
logger2.debug("this is a debug message for logger2") # so logger2 will not show debug as its level is set above debug which is warning 
logger2.warning("this is a warning message")
logger2.error("this is a error message")
```
