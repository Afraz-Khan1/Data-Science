```python
# but when reading from file there is chances weather there is a missing value or not.
# so how to check weather there are missing values or not
# .isnull method will tell you with a true keyword that here is the missing value
# and if the value is present in the block there will be false
data.isnull() 
```

```python
#however it is hard to find the true 
# there are different ways to see like
# .isnull().sum() this whole function will sum the missing value specific to each column 
data.isnull().sum() 
```

```python
data.isnull().any() # .isnull().any() will show which column has the missing values by showing true keyword on the same column
# this is usefill .isnull().any() as it tells you which column has the missing values.
# later you can use that column name and fill missing ones with other value
```

```python
# to fill those missing spaces rather with zero or the mean of that whole coloumn.
# we adding another column with missing blocks filled with whole block mean value.
#data['Value'] means this value column fillna with data['value].mean(). mean whole value column. 
# and you can see value column where there is NAN means in value_filledna in the same line there will be missing box filled with value column mean.
data["value_filledna"]=data['Value'].fillna(data['Value'].mean())  
```

```python
data
```

```python
data['Sales_filledna']=data['Sales'].fillna(data['Sales'].mean())
# and you can see sales column where there is NAN means in sales_filledna in the same line there will be missing box filled with sales column mean.
```

```python
data
```

```python
# so now removing sales and value column as we have added corrected ones
data.drop('Value',axis=1,inplace=True) # as we removing column we need to add axis =1 so that whole column will be removed.
data.drop('Sales',axis=1,inplace=True) # as we removing column we need to add axis =1 so that whole column will be removed.
```

```python
data # :) successfully removed.
```

```python
# renaming the column name
data=data.rename(columns={'Date':'Sales_data'}) # it should be : as they are key values pairs in dict.
data
```

```python
# One more thing if we apply any operation on the value and sales columns that we removed earlier it will show error as they have missing values. so we created new ones with all missing boxes fiiled so we can use them in operations easily.
# changing data types.
# changing data types of all the values in column value_filldna
data['value_filledna']=data['value_filledna'].astype(int) # inorder to save those all changed data types in the column we must assign this operation to the column.
```

```python
data
```

```python
# changing values by using any function or lambda function your choice
data['value_filledna_new']=data['value_filledna'].apply(lambda x:x*2) # this apply function can be used to apply funciton to this column only.
# so we added a new column with function applied answer of value_filledna
```

```python
data
```

```python
# aggregating : aggregating means grouping with respect to particular thing. here is columns.
data.groupby('Product')['value_filledna'].sum() # when you code like this typing product or value_filledna the IDE will not auto pick idk why.
# so what happened above is you selected a column product and grouped all the same values in product with there values in value_filledna summed them. 
```

```python
data.groupby(['Product','Region'])['value_filledna'].sum()
```

```python
ope=data.groupby('Region')['value_filledna'].agg(['sum','mean','count']) # agg is aggregation builtin method that is used for using these function like sum, mean,count.
ope
```

```python
#merging dataframes
import pandas as pd
df1=pd.DataFrame({'Key':['A','B','C','D'],'Value1':[1,2,3,4]})
df2=pd.DataFrame({'Key':['A','B','C','E'],'Value2':[5,6,7,8]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='Key',how='inner')
```

```python
pd.merge(df1,df2,on='Key',how='outer')
```

```python
pd.merge(df1,df2,on='Key',how='left') # left means df1 will be given priority that all the keys of df2 will be present in df1 than it will merge or else it will elave that one
```

```python
pd.merge(df1,df2,on='Key',how='right') # as d was not present in key of df1 so this merge didn't pick d one.
```
