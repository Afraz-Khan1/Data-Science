```python
# series in pandas.
# it can hold any data type.
import pandas as pd
data=[1,2,3,4,5]
series=pd.Series(data)
print("Series: \n",series) # 0 1 2 3 4 5 are the indexing, 1 2 3 4 5 are now in series.
```

```python
# series in dictionary
dict={'a':1,'b':2,'c':3,'d':4,'e':5,'f':6} # now a b c.. are the index of numbers
series_dict=pd.Series(dict)
print("series of dict:\n",series_dict)
print("Type:",type(series_dict))
```

```python
data=[1,2,3]
index=[10,20,30] # setting manual index is also possible
pd.Series(data,index=index)
```

```python
# jumping towards creating dataframe through dictionary of list.
dict1={'Name':["afraz","abdullah","faiz"],"Age":[19,25,20],"City":["karahi","karachi","chau"],"gender":["male","male","male"]}
df=pd.DataFrame(dict1)
print(df)
print(type(df))
import numpy as np
# we can also use numpy for this to see in array form
nn=np.array(df)
print("\nUsing numpy arrays:\n",nn)
```

```python
# same thing but doing it with a list of dictionaries
lsst=[{'Name':"afraz","Age":19,"City":"Karachi","Gender":"Male"},
      {"Name":"Abdullah","Age":25,"City":"Karachi","Gender":"Male"},
      {"Name":"Faiz","Age":25,"City":"Chou","Gender":"Male"}]
dflsst=pd.DataFrame(lsst)
print("Data Frame As List of Dictionaries:\n",dflsst)
print("\nType:\n",type(dflsst))
nn=np.array(dflsst)
print("\nUsing numpy arrays:\n",nn)
```
