```python
import seaborn as sns
tip=sns.load_dataset('tips') # this tip is builtin dataset in sns github. we just read it
tip
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tip,color='blue') # here data=tip tells the seaborn where to find these two columns in dataframe or else seaborn can't find
```

```python
# we can also import matplotlib to do other operation like changing the titles
import matplotlib.pyplot as plt
sns.lineplot(x='size',y='total_bill',data=tip)
plt.title('Total_bill vs Size')
```

```python
# bar plot
import seaborn as sns
sns.barplot(x='day',y='total_bill',data=tip,color='teal')
```

```python
# box plot
sns.boxplot(x='day',y='total_bill',data=tip)
```

```python
# violin plot
sns.violinplot(x='day',y='total_bill',data=tip)
```

```python
# histogram plot
sns.histplot(tip['total_bill'],bins=10,kde=True) # kde is curve
```
