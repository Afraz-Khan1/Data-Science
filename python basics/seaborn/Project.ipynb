<h1>(Heart Attack Dataset) data analysis using pandas and numpy and visualization using seaborn and matplotlib</h>

```python
import pandas as pd
df=pd.read_csv("heart_attack_prediction_dataset.csv")
df=df.rename(columns={"Sex":"Gender"})
df.head(50)
```

```python
# Heart attack risk on the basis of gender and physical acitvity factor.
d1=df.groupby(["Gender","Physical Activity Days Per Week"])["Heart Attack Risk"].mean().reset_index()
d1
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
sns.barplot(x="Physical Activity Days Per Week",y="Heart Attack Risk",hue="Gender",legend="auto",palette="Set2",data=d1)
plt.title("Heart attack vs Physical Activity")
#insight: physical activity makes heart attack risk decrease for individuals. 
```

```python
d1=df.groupby(["Gender","Sleep Hours Per Day"])["Heart Attack Risk"].mean().reset_index()
d1
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
d1=df.groupby(["Gender","Sleep Hours Per Day"])["Heart Attack Risk"].mean().reset_index()
sns.scatterplot(x="Sleep Hours Per Day",y="Heart Attack Risk",hue="Gender",legend="auto",data=d1)
plt.title("Sleep Hours Effect On Heart")
# the insight we got is having a long sleep in both male and female heart attack risk was found lesser .
```

```python
d2=df.groupby(["Family History"],as_index=False)["Heart Attack Risk"].mean()
d2
```

```python
sns.barplot(x="Family History",y="Heart Attack Risk",color="lightcoral",data=d2)
plt.title("Family History Of Heart Attack")
# Insight: family history heart attack risk doesn't meant to necessarily increase the individual's rate of heart attack. 
```

```python
# smokeing causes heart attack?
d3=df.groupby(["Smoking","Cholesterol","Diabetes"],as_index=False)["Heart Attack Risk"].mean()
d3
```

```python
d3=df.groupby(["Cholesterol","Gender"],as_index=False)["Heart Attack Risk"].mean()
d4=df.groupby(["Diabetes","Smoking"],as_index=False)["Heart Attack Risk"].sum()
d5=df.groupby(["Blood Pressure","Cholesterol"],as_index=False)["Heart Attack Risk"].mean()


plt.figure(figsize=(16,12))
plt.subplot(2,3,1)
sns.scatterplot(x="Cholesterol",y="Heart Attack Risk",hue="Gender",data=d3,palette="Set1")
plt.title("Cholesterol Factor")

plt.subplot(2,3,2)
sns.barplot(x="Diabetes",y="Heart Attack Risk",hue="Smoking",data=d4,palette="Set2") 
plt.title("Diabetes Factor")

plt.subplot(2,3,4)
sns.histplot(x="Blood Pressure",bins=20,color="purple",data=d5) # only x used in graph.
plt.title("Blood Pressure Factor")

# Insight: High cholesterol and diabetes have critical impact on heart.
# As if we consider cholestrol high in individual's it was observed females have more heart attack cases over males.
# whereas considering smoking with diabetes it came up with increased heart attack risk results. 
```
