```python
# use dictionaries to count the frequency of elements 
numbers=[1,2,2,3,3,3,4,4,4,5,5,5,5]
frequency={}
for number in numbers:
    if number in frequency:
        frequency[number]+=1 # increase the count of number if present in frequency or simply edits the value of the number
    else:
        frequency[number]=1 # initializes the number in frequency if not present
print(frequency)
```

```python
# there are function parameters and default parameters (parameters = variable name in fucntion definition)
# function parameters argument are given from where it is called. and default parameters argument are given in the function definition.
# now there are variable length arguments
# 1) positional arguments
# 2) keyword arguments.
#example of both 
def cal(*pos):
    print("positional arguments: ") ## the positions are given of the values it is like an array of elements *pos
    for val in pos:
        print(val)
cal(1,2,3,4,5,6,"afraz",3.5)
## keyword arguments: means there will be given values with there key so this **args points to both the key and the value
def cal2(**args):
    print("\nkeyword arguments:")
    for key,value in args.items(): ## **args is just like dictionary we unpack the items.
        print(f"{key}:{value}")
cal2(calculate=123,name="afraz",age=19,degree="not achived yet")
print(type('?'))

```

```python
# pass word strength checker:
def check_strength(password):
    if len(password)<=5:
        return print("password length should be greater than 4")
    if not any(i.islower() for i in password): ## this for loop is a generator for loop which checks one by one and stops if it founds one lowercase. it skips if there is uppercase letter
       ## if there is no lowercase condition gives false and not(false)true false returned. this same happens in all.
       # here atleast one should be lower in order to make password multitype
        return False
    if not any(i.isupper() for i in password): #generator stops as one upper is found but jumps to next for loop as not make it false.
        return False
    if not any(i.isdigit() for i in password):
        return False
    if not any(i in '&*()-^%$#@!' for i in password):
        return False
    return True
print(check_strength("AF!rAZ123*"))  
print(check_strength("afrazP"))
```

```python
# make a function which tracks the price of products bought.
def price_total(cart):
    total=0
    for item in cart:
        total+=item['price']*item['quantity']
    return total

cart=[
    {'name':'apple','price':10,'quantity':4},
    {'name':'mango','price':10,'quantity':9},
    {'name':'banana','price':10,'quantity':10}
]

total=price_total(cart)
print(total)
```

```python
# make a fucntion to check if srting is palindrome or not
def cal_palin(string):
    return string==string[::-1]
print(cal_palin("1badfdab1"))
```

```python
# make a function that will calculate the factorial using recursion
def cal_fat(number):
    if number==1:
        return 1
    return cal_fat(number-1)*number # for 5 : first step: number 5 is multiplied after getting factorial of 5-1. second: number 4 is multiplied after getting factorial of 4-1...
fact=cal_fat(4)
print(fact)
```

```python
# count lines words and characters in the file data.

 # split is used to seperate the words when encounter a white space. works only with strings
def count_lines_words_characters(file):
  with open(file,mode='r') as source:
    lines=source.readlines() # read all the lines at once
    count_lines=len(lines)
    count_word=sum(len(line.split()) for line in lines) # so this line in lines : line is the single line which is splitted into words based on whitespace and its length is calcualted and is summed with next line word count in the same way.
    count_characters=sum(len(line) for line in lines)  # here simple: line is the single line whos length is calculated which directly gives the characters count and sum() is used to sum it with next lines characters count.
  return count_lines,count_word,count_characters
  
file_path='file1.txt'
count1,count2,count3=count_lines_words_characters(file_path)
print(f"Total lines: {count1},Total words: {count2},Total characters: {count3}")

    
```

```python
# w+ mode is used for reading and writing both but it has special ability to create the file if it doesn't exist or append data if it exist.
# but if you overwrite the data then when reading you must make sure the cursor is in the beginning or else file will not read. use file.seek(0)
with open('file3.txt',mode='w+')as file: # but this w+ dont append it overwrite
    while True:
        line=input("Enter the data to add into file: Enter on empty line to stop: ")
        if line=="":
            break
        file.write(line +'\n')
    file.seek(0)
    content=file.read()
    print(content)
```
